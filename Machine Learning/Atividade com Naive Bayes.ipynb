```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
```

```python
dados = pd.read_csv('soybean.csv', sep= ',')
dados
```

```python
previsores = dados.iloc[:,0:35].values
classe = dados.iloc[:,35].values
previsores
```

```python
labelencoder = LabelEncoder()
for x in range(35):
    previsores[:,x] = labelencoder.fit_transform(previsores[:,x])
```

```python
x_treino, x_teste,y_treino, y_teste = train_test_split(previsores, classe, test_size= 0.3, random_state= 0)
```

```python
modelo = GaussianNB()

modelo.fit(x_treino, y_treino)
```

```python
previsao = modelo.predict(x_teste)
previsao
```

```python
taxa_acerto = accuracy_score(y_teste, previsao)
taxa_acerto
```

modelo apresentou acerto de 85,85%
