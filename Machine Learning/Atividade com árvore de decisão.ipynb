```python
import pandas as pd #biblioteca para importação dos dados em csv, criação do dataframe e manipulação de arrays
from sklearn.model_selection import train_test_split #função para a separaçãos dos dados do treino e teste
from sklearn.preprocessing import LabelEncoder # função para converter os previsores de strings para números
from sklearn.metrics import accuracy_score #pacalculo da taxa de acerto do modelo
from sklearn.tree import DecisionTreeClassifier #estipular o número de árvores que será criado para definir a com melhores resultados
from sklearn.ensemble import RandomForestClassifier #treino do modelo de árvores de decisão
```

```python
dados = pd.read_csv('soybean.csv', sep= ',')
dados
```

```python
previsores = dados.iloc[:,0:35].values
classe = dados.iloc[:,35].values
previsores
```

```python
labelencoder = LabelEncoder()
for x in range(35):
    previsores[:,x] = labelencoder.fit_transform(previsores[:,x])
```

```python
x_treino, x_teste,y_treino, y_teste = train_test_split(previsores, classe, test_size= 0.3, random_state= 0)
```

```python
floresta = RandomForestClassifier(n_estimators= 200)
floresta.fit(x_treino, y_treino)
```

```python
previsoes = floresta.predict(x_teste)
taxa_acerto = accuracy_score(y_teste, previsoes)
taxa_acerto
```

Modelo aresta taxa de acerto de 95,60%
