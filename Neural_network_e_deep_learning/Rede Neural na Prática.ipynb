Redes neurais artificiais com keras

```python
# importação das biblotecas
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
# utilizado a biblioteca Keras por ser uma abstração sobre a biblioteca tensorflow, facilitando seu uso
from keras.models import Sequential
from keras.layers import Dense
from keras.utils import np_utils
import numpy as np
```

```python
#carregar dados iris
base = datasets.load_iris()
#divisão dos dados em variáveis independentes e dependentes
previsores = base.data
classe = base.target
classe
```

```python
#organicar cada classe em uma coluna para atender a rede neural
classe_dummy = np_utils.to_categorical(classe)
classe_dummy
```

```python
#divisão de dados em treino e teste
x_treinamento, x_teste, y_treinamento, y_teste = train_test_split(previsores, classe_dummy, test_size= 0.3, random_state= 0)
```

```python
#criação da estrutura da rede neural com classe Sequential
modelo = Sequential()
#primeira camada oculta, 5 neurnonios, 4 neuronio de entrada
modelo.add(Dense(units= 5, input_dim = 4)) #pode ser elevado o número de neurônios, porém os de entrada devem ser penas quatro por termos quatro informaões de entrada
#segunda camada oculta
modelo.add(Dense(units= 4))
#terceira camada oculta
#função softbox porque temos um problema de classificação com mais de duas classes
#é gerada um probabilidade em cada neunrônio
modelo.add(Dense(units= 3, activation = 'softmax'))
```

```python
#exibir resultado de como ficou organizado a rede
modelo.summary()
```

```python
#configuração de parâmetros da rede neural (adam = algorítmo para tualizaros pesos e Loss = cálculo de erro)
modelo.compile(optimizer= 'adam', loss= 'categorical_crossentropy', metrics= ['accuracy'])
#treianmento, dividindo a base de treino em porções para validação (validation_data)
modelo.fit(x_treinamento, y_treinamento, epochs= 1000, validation_data= (x_teste, y_teste))
```

No Epoch 1000/1000 é possível ver que o modelo com os dados de treino (accuracy) apresenta uma performance de 0.9905 e com os dados de teste (val_accuracy) 0.9778

```python
#previsoes e midar a variável para True ou false de acordo com o threshold 0.5
previsoes = modelo.predict(x_teste)
previsoes = (previsoes > 0.5)
previsoes
```

```python
#como é um problmea com três saídas, precisamos buscar a posição que possui o maior valor (são retornados três valores)
y_teste_matrix = [np.argmax(t) for t in y_teste]
y_previsao_matrix = [np.argmax(t) for t in previsoes]
```

```python
#geração de matrix de confusão
confusao = confusion_matrix(y_teste_matrix, y_previsao_matrix)
confusao
```
