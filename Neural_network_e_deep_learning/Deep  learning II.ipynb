```python
import pandas as pd
from keras.models import Sequential
from keras.layers import Dense
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.compose import make_column_transformer
```

```python
dataset = pd.read_csv('Credit2.csv', sep= ';')
dataset
```

```python
#separaÃ§Ã£o de dados em suas variÃ¡veis
x = dataset.iloc[:,1:10].values # coluna ID ignorada por nÃ£o ter valores Ãºteis
y = dataset.iloc[:,10].values

x
```

```python
#transformar a coluna "Checking_status" em nÃºmeros de 0 a 3
labelencoder = LabelEncoder() # utilizado a funÃ§Ã£o LabelEncoder por se tratar de classificaÃ§Ã£o sequencial 
x[:,0] = labelencoder.fit_transform(x[:,0])
x
```

```python
# traformaÃ§Ã£o dos dados classificatÃ³rios "credit_history" em nÃºmeros
#utilizada a funÃ§Ã£o OneHotEncoder por se tratar de de classificaÃ§Ã£o categÃ³rica, entÃ£o cada caracterÃ­stica serÃ¡ colocada em uma coluna
onehotencoder = make_column_transformer((OneHotEncoder(categories= 'auto', sparse= False), [1]), remainder= 'passthrough')
x = onehotencoder.fit_transform(x)
x
```

```python
# exclusÃ£o de uma das colunas criadas pelo processo ONeHotEncoder para evitar erros gerados pela correlaÃ§Ã£o delas
x = x[:,1:]
x
```

```python
#tranformaÃ§Ã£o dos das da classe em valoes numÃ©ricos
labelencoder_y = LabelEncoder()
y = labelencoder_y.fit_transform(y)
y 
```

```python
x_train, x_test, y_train, y_teste = train_test_split(x, y, test_size= 0.2, random_state= 0)
print(len(x_train), len(x_test), len(y_train), len(y_teste)) # visualizaÃ§Ã£o do tamanho das amostras de treino e teste
```

```python
# normalizaÃ§Ã£o do intervalo de variÃ¡veis transformando todos os dados em uma mesma escala
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.fit_transform(x_test)
x_test
```

```python
classifier = Sequential()
classifier.add(Dense(units= 6, kernel_initializer= 'uniform', activation= 'relu', input_dim= 12))
classifier.add(Dense(units= 6, kernel_initializer= 'uniform', activation= 'relu'))
classifier.add(Dense(units= 1, kernel_initializer= 'uniform', activation= 'sigmoid'))
classifier.compile(optimizer= 'adam', loss= 'binary_crossentropy', metrics= ['accuracy'])
classifier.fit(x_train, y_train, batch_size= 10, epochs= 100)
```

```python
#previsÃ£o com os dados de teste
y_pred = classifier.predict(x_test)
#resultado Ã© entregue em probabilidades, entÃ£o serÃ¡ convertido para true ou false
y_pred = (y_pred > 0.5)
y_pred
```

```python
confusao = confusion_matrix(y_teste, y_pred)
confusao
```
