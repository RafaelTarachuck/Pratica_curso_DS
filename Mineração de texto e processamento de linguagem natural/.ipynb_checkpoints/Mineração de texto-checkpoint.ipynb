```python
import matplotlib.pyplot as plt
import nltk
nltk.download('stopwords')
from nltk.corpus import PlaintextCorpusReader
from nltk.corpus import stopwords
from matplotlib.colors import ListedColormap
from wordcloud import WordCloud
import string
```

```python
#criação de um corpus lendo textos do disco
corpus= PlaintextCorpusReader('Arquivos', '.*', encoding= 'ISO-8859-1')
```

```python
#leitura dos arquivos de disco. Percorrer os arquivos e mostrar o nome dos primeiros 100 arquivos
arquivos= corpus.fileids()
#primeiro arquivo
arquivos[0]
```

```python
# zero até 107
arquivos[0:10]
```

```python
#imprimir todos os nomes
for a in arquivos:
    print(a)
```

```python
#acessoa text d primeiro arquivo
texto= corpus.raw('1.txt')
texto
```

```python
#acesso a todas as palavras de todos os aquivos
todo_texto= corpus.raw()
#todo_texto
```

```python
palavras=corpus.words()
palavras[170]
```

```python
#quantidade
len(palavras)
```

```python
#usando nltk se obtem as stop words em inglês
stops = stopwords.words('english')
stops
```

```python
#definição das cores que serão utilizadas na nuvem de palavras
mapa_cores= ListedColormap(['orange', 'green', 'red', 'magenta'])
#criação da nivem de palavras com no máximo 100 palavras e utilizando as stop word
nuvem= WordCloud(background_color = 'white', colormap= mapa_cores, stopwords= stops, max_words= 100)
#criando e visualização de nuvem de palavras
nuvem.generate(todo_texto)
plt.imshow(nuvem)
```

```python
#nova lista sem stop word
palavras_semstop = [p for p in palavras if p not in stops]
len(palavras_semstop)
```

```python
#remoção de pontuação. Lista sem pontuação ou stop word
palavras_sem_pontuacao = [p for p in palavras_semstop if p not in string.punctuation]
len(palavras_sem_pontuacao)
```

```python
#calculo frquancia das paavras e visualização das masi comuns
frequencia= nltk.FreqDist(palavras_sem_pontuacao)
frequencia
```

```python
#mais comuns
mais_comuns = frequencia.most_common(100)
mais_comuns
```
