```python
from igraph import Graph
from igraph import plot
```

```python
grafo5 = Graph(edges= [(0,1), (2,3), (0,2), (0,3)], directed= True)
grafo5.vs['label']= ['Fernando', 'Pedro', 'Jose', 'Antonio']
grafo5.vs['peso']= [40, 30, 30, 25]
grafo5.es['Tipo relação'] = ['Amigo', 'Inimigo', 'Inimigo','Amigo']
grafo5.es['weight'] = [1,2,1,3]
```

```python
for x in grafo5.vs:
    print(x)
```

```python
#definição de cor para o vértices
grafo5.vs['cor']= ['blue', 'red', 'yellow', 'green']
plot(grafo5, bbox= (0,0,300,300), vertex_color= grafo5.vs['cor'])
```

```python
#definição de peso para as arestas
plot(grafo5, bbox=(300,300), edge_width= grafo5.es['weight'], vertex_color= grafo5.vs['cor'])
```

```python
#peso para o vértices
plot(grafo5, bbox= (300,300), vertex_size = grafo5.vs['peso'], edge_width= grafo5.es['weight'], vertex_color= grafo5.vs['cor'])
```

```python
#curvatura
plot(grafo5, bbob=(300, 300), vertex_size= grafo5.vs['peso'], edge_width= grafo5.es['weight'], vertex_color= grafo5.vs['cor'], edge_curved= 0.4)
```

```python
#formato
plot(grafo5, bbox=(300,300), vertex_size= grafo5.vs['peso'], edge_width= grafo5.es['weight'], vertex_color= grafo5.vs['cor'], edge_curved=0.4, vertex_shape= 'square')
```
